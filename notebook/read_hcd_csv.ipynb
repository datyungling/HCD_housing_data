## Data Cleaning

```python
# Define dtypes for each column per table
COLUMN_TYPES = {
    'hcd_table_a': {
        'numeric': [
            'VLOW_INCOME_DR',
            'VLOW_INCOME_NDR',
            'LOW_INCOME_DR',
            'LOW_INCOME_NDR',
            'MOD_INCOME_DR',
            'MOD_INCOME_NDR',
            'ABOVE_MOD_INCOME',
            'TOT_PROPOSED_UNITS',
            'TOT_APPROVED_UNITS',
            'TOT_DISAPPROVED_UNITS',
            'YEAR'
        ],
        'date': [
            'APP_SUBMIT_DT',
        ],
    },
    'hcd_table_a2': {
        'numeric': [
            # Entitlements
            'VLOW_INCOME_DR',
            'VLOW_INCOME_NDR',
            'LOW_INCOME_DR',
            'LOW_INCOME_NDR',
            'MOD_INCOME_DR',
            'MOD_INCOME_NDR',
            'ABOVE_MOD_INCOME',
            'NO_ENTITLEMENTS',
            # Building permits
            'BP_VLOW_INCOME_DR',
            'BP_VLOW_INCOME_NDR',
            'BP_LOW_INCOME_DR',
            'BP_LOW_INCOME_NDR',
            'BP_MOD_INCOME_DR',
            'BP_MOD_INCOME_NDR',
            'BP_ABOVE_MOD_INCOME',
            'NO_BUILDING_PERMITS',
            # Certificates of occupancy
            'CO_VLOW_INCOME_DR',
            'CO_VLOW_INCOME_NDR',
            'CO_LOW_INCOME_DR',
            'CO_LOW_INCOME_NDR',
            'CO_MOD_INCOME_DR',
            'CO_MOD_INCOME_NDR',
            'CO_ABOVE_MOD_INCOME',
            'NO_OTHER_FORMS_OF_READINESS',
            # Other
            'EXTR_LOW_INCOME_UNITS',
            'DEM_DES_UNITS',
            'DENSITY_BONUS_TOTAL',
            'DENSITY_BONUS_NUMBER_OTHER_INCENTIVES',
        ],
        'date': [
            'ENT_APPROVE_DT1',
            'BP_ISSUE_DT1',
            'CO_ISSUE_DT1',
        ],
    },
}



def apply_types(df: pd.DataFrame, table_name: str) -> pd.DataFrame:
    type_map = COLUMN_TYPES.get(table_name, {})

    for col in type_map.get('numeric', []):
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')

    for col in type_map.get('date', []):
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors='coerce')

    return df


# Load and convert

table_a = apply_types(table_a, 'hcd_table_a')
table_a2 = apply_types(table_a2, 'hcd_table_a2')
```

```python
table_a.UNIT_CAT.unique()
```

```python
VALID_CATS = ['SFD', 'SFA', '2 to 4', '5+', 'ADU', 'MH']
```

# Analysis: Permitted Units by Structure Type (pg 9 of dashboard)

```python
# Select one year reported from APR dashboard for comparison
df_hcd_dash = pd.read_csv('../data/hcd_permits_dash.csv')
df_hcd_dash_2023 = df_hcd_dash[df_hcd_dash['Year'] == 2023].transpose().rename(columns={10: 'dashboard permitted units'})
df_hcd_dash_2023 = df_hcd_dash_2023.drop(index='Year')
df_hcd_dash_2023
```

```python
# Add year columns for easier grouping
table_a['APP_SUBMIT_DT_YEAR'] = table_a['APP_SUBMIT_DT'].dt.year
table_a2['BP_ISSUE_DT1_YEAR'] = table_a2['BP_ISSUE_DT1'].dt.year
```

```python
# For table_a, getting YEAR (report year)= 2023 and APP_SUBMIT_DT_YEAR (permit submit year) = 2023, 
# then group by UNIT_CAT and sum TOT_PROPOSED_UNITS and TOT_APPROVED_UNITS. 
# Reindex to show all categories in VALID_CATS, and add a Total row at the end.

table_a_2023 = table_a[
    (table_a['YEAR'] == 2023) & (table_a['APP_SUBMIT_DT_YEAR'] == 2023)
].reset_index(drop=True)

print(len(table_a_2023))

#just looking at the APP SUBMIT DT YEAR yields a larger number of permitted approved units
table_a_2023 = table_a[table_a['APP_SUBMIT_DT_YEAR'] == 2023]

result_2023 = (
    table_a_2023
    .groupby('UNIT_CAT')[['TOT_PROPOSED_UNITS', 'TOT_APPROVED_UNITS', 'TOT_DISAPPROVED_UNITS']]
    .sum()
    .reindex(VALID_CATS)
)
print(result_2023.to_string())
```

```python
#For table_a2, getting YEAR (report year)= 2023 and BP_ISSUE_DT1_YEAR (building permit issue year) = 2023,
# then group by UNIT_CAT

table_a2_2023 = table_a2[
    (table_a2['YEAR'] == 2023) & (table_a2['BP_ISSUE_DT1_YEAR'] == 2023)
].reset_index(drop=True)    

print(len(table_a2_2023))

result_2023_a2 = (
    table_a2_2023
    .groupby('UNIT_CAT')[['NO_BUILDING_PERMITS', 'NO_OTHER_FORMS_OF_READINESS', 'NO_ENTITLEMENTS']]
    .sum()
    .reindex(VALID_CATS)
)

print(result_2023_a2.to_string())
```

```python
#For table_a2, trying just BP_ISSUE_DT1_YEAR (building permit issue year) = 2023,
# then group by UNIT_CAT
# this produces a much larger number of permitted units than the previous filter

table_a2_2023 = table_a2[table_a2['BP_ISSUE_DT1_YEAR'] == 2023] 

print(len(table_a2_2023))

result_2023_a2 = (
    table_a2_2023
    .groupby('UNIT_CAT')[['NO_BUILDING_PERMITS', 'NO_OTHER_FORMS_OF_READINESS', 'NO_ENTITLEMENTS']]
    .sum()
    .reindex(VALID_CATS)
)

print(result_2023_a2.to_string())
```

```python
# Compare all the data
output = pd.concat(
    [df_hcd_dash_2023, result_2023, result_2023_a2],
    axis=1,
    keys=['dashboard', 'table_a', 'table_a2']
)

print(output)
```

```python
import matplotlib.pyplot as plt

cols = ['dashboard permitted units', 'TOT_APPROVED_UNITS', 'NO_BUILDING_PERMITS']

plot_df = output.loc[:, [('dashboard', 'dashboard permitted units'), 
                          ('table_a', 'TOT_APPROVED_UNITS'), 
                          ('table_a2', 'NO_BUILDING_PERMITS')]]

plot_df.columns = ['Dashboard Permitted', 'TOT_APPROVED_UNITS', 'NO_BUILDING_PERMITS']

ax = plot_df.plot(kind='bar', figsize=(10, 6))
ax.set_title('Permitted Units by UNIT_CAT (2023)')
ax.set_xlabel('UNIT_CAT')
ax.set_ylabel('Units')
ax.tick_params(axis='x', rotation=45)

for container in ax.containers:
    ax.bar_label(container, fmt='{:,.0f}', label_type='edge', padding=3, fontsize=8)

plt.tight_layout()
plt.show()

```
